```python
fig, axs = plt.subplots(1, 2, figsize=(15,7))

#Plot 1 Tution_USD
sns.histplot(data=df,x='gender',bins=20,kde=False,color='g', ax = axs[0])
axs[0].set_title('Distribution of Gender')
axs[0].set_xlabel('Tuition (USD)')
axs[0].set_ylabel('Frequency')

#Plot 2 Country
sns.countplot(data=df, x='country', color='b', ax=axs[1])
axs[1].set_title(' Distribution of Country')
axs[1].set_xlabel('Country')
axs[1].set_ylabel('Frequency')
axs[1].tick_params(axis='x', rotation=45)
plt.show()
```

#### 4.14 Exploring the Average Tuition by Country

```python

avg_mood_country = df.groupby('country')['mood'].mean().sort_values(ascending=False)

#Plotting
plt.figure(figsize=(30,20))
sns.barplot(x=avg_mood_country.values,y=avg_mood_country.index,palette='viridis')
plt.title('Mood by Country')
plt.xlabel('Mood')
plt.ylabel('Country')
plt.show()
```

### 4.18 Correlation

```python
corr = df[['age', 'mood', 'stress_level', 'sleep_hours', 'screen_time_hours', 'used_ai_today', 'journaled_today', 'meditated_today', 'exercised_today', 'social_interaction_rating', 'support_feeling']].corr()

sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

## Insights

Strong Negative Correlation: mood vs. stress_level is about -0.926, indicating that as stress increases, mood tends to decrease.
